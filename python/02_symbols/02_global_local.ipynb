### Global 

```python
def foo():
    print("inside foo", b)


b = 9
foo()
print("main", b)
```

```python
b = 9


def foo():
    print("inside foo", b)


foo()
print("main", b)
```

###  Local

```python
def bar():
    y = 7
    print("inside bar", y)


bar()
print("main", y) # expect error
```

### Pay attention!

```python
b = 9


def magic():
    y = 7
    print("inside magic", y, b)
    b = 6


print("main", b)
magic()
```

```python
b = 9


def magic_local():
    y = 7
    b = 6
    print("inside magic_local", y, b)


magic_local()
print("main", b)
```

### `global` keyword

```python
b = 9


def magic_global():
    global b  # <----
    y = 7
    print("inside magic_local", y, b)
    b = 6


magic_global()
print("main", b)
```
