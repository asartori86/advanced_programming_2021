```python
l[2].append(7)
print(l)
print(o)
```

```python
import copy
```

```python
o = copy.copy(l)
l[2].append(888)
print(l)
print(o)
```

```python
o = copy.deepcopy(l)
l[2].append(3333)
print(l)
print(o)
```

```python
def augmented(ll):
    ll += [99]


augmented(l)
print(l)
```

```python
def assignment(ll):
    ll = [99]


assignment(l)
print(l)
```
